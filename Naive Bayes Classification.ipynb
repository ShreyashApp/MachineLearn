```python
file = open("winequality-red.csv","r")
data = []
for line in file:
    a=line.split(",")
    l2=[]
    for j in range(0,len(a),1):
        l2.append(float(a[j]))
    data.append(l2)
file.close()
rows = len(data)
cols=len(data[0])
n=[]
n.append(0)
n.append(0)
avgrat = 0
trainlabels = {}
for i in range(int(rows*3/5),rows,1):
    if data[i][cols-1] in trainlabels:
        trainlabels[data[i][cols-1]]+=1
    else:
        trainlabels[data[i][cols-1]]=1
for rw in range(0,int(rows*3/5),1):
    avgrat+=data[rw][cols-1]
avgrat/=rows
for rw in range(0,int(rows*3/5),1):
    if(data[rw][cols-1]<avgrat):
        n[0]+=1
    else:
        n[1]+=1
m0=[]
temps0=[]
s0=[]
for j in range(0,cols-1,1):
    m0.append(0)
    s0.append(0)
    temps0.append(0)
m1=[]
temps1=[]
s1=[]
for j in range(0,cols-1,1):
    m1.append(0)
    s1.append(0)
    temps1.append(0)
for i in range(0,int(rows*3/5),1):
    if(data[i][cols-1]<avgrat):
        for j in range(0,cols-1,1):
            m0[j]=m0[j]+data[i][j]
    else:
        for j in range(0,cols-1,1):
            m1[j]=m1[j]+data[i][j]
for j in range(0,cols-1,1):
    m0[j]=m0[j]/n[0]
    m1[j]=m1[j]/n[1]
for i in range(0,int(rows*3/5),1):
    if(data[i][cols-1]<avgrat):
        for j in range(0,cols-1,1):
            temps0[j]+=(data[i][j]-m0[j])**2
    else:
        for j in range(0,cols-1,1):
            temps1[j]+=(data[i][j]-m1[j])**2
for j in range(0,cols-1,1):
    s0[j]=(temps0[j]/n[0])**(0.5)
    s1[j]=(temps1[j]/n[1])**(0.5)
#print(m0,m1,s0,s1)
a=0
b=0
ck=0
d=0
file = open("prediction","w")
file.write('The dataset is divided into two. One to train and one set to validate. The ratio is 3:2'+'\n')
for i in range(int(rows*3/5),rows,1):
        d0=0
        d1=0
        if(data[i][cols-1]<avgrat):
            c=0
        else:
            c=1
        for j in range(0,cols-1,1):
            d0+=((data[i][j]-m0[j])/s0[j])**2
            d1+=((data[i][j]-m1[j])/s1[j])**2
        if(d0<d1):
            file.write('0 '+str(i)+'\n')
            ('0 '+str(i))
            cp=0
        else:
            file.write('1 '+str(i)+'\n')
            print('1 '+str(i))
            cp=1
        if(c==0 and cp==0):
            a+=1
        if(c==0 and cp==1):
            b+=1
        if(c==1 and cp==0):
            ck+=1
        if(c==1 and cp==1):
            d+=1
error = 0.5*(a/(a+b)+d/(c+d))
file.write("error: "+str(error))
print(str(a)+"\t"+str(b)+"\n"+str(c)+"\t"+str(d))
print(str(error))
file.close()
```
